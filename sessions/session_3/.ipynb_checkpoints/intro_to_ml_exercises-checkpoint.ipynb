# Introduction to Machine Learning and Data Science: Machine Learning basics with Diabetes and Aquatic Toxicity datasets

This notebook contains exercises for an introductory course on Machine Learning and Data Science, focusing on a regression and a classification problem with the QSAR Aquatic Toxicity and the Pima Indians Diabetes datasets, respectively.

We will cover essential steps in understanding, cleaning, transforming, and analyzing the data using pandas and matplotlib/seaborn. The exercises are divided into basic, intermediate, and advanced levels.

## Setup

We already used the Diabetes dataset last week, but here are the names of the columns, since it comes without them:
1. Pregnancies
2. Glucose
3. BloodPressure
4. SkinThickness
5. Insulin
6. BMI
7. DiabetesPedigreeFunction
8. Age
9. Outcome (0 for no diabetes, 1 for diabetes)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# URL for the Pima Indians Diabetes dataset
url_classification = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names_classification = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']

# Load the dataset
diabetes_df = pd.read_csv(url_classification, names=names_classification)

# Display the first few rows to confirm loading
diabetes_df.head()
```

For the regression dataset, we will use the QSAR Aquatic Toxicity dataset, which was used to develop a quantitative structure-activity relationship models to predict aquatic toxicity towards the fish Pimephales promelas (fathead minnow). LC50, which is the concentration that causes death in 50% of test Daphnia Magma over a test duration of 48 hours, is the target.

```python
url_regression = "https://raw.githubusercontent.com/readytensor/rt-datasets-regression/refs/heads/main/datasets/processed/aquatic_toxicity/aquatic_toxicity.csv"

aquatic_toxicity_df = pd.read_csv(url_regression)
aquatic_toxicity_df.head()
```

### Exercise 1: Splitting the Data

1. Separate features (`X`) and target (`y`).
2. Perform the data splitting with the `sklearn.model_selection.train_test_split`. Get the training, validation and test sets with a 60/20/20  split. **Only for classification dataset**: perform the splitting with the `stratify` option set to yes. It attempts to keep the percentages of the classes in the different sets.
3. Print the shape of both sets, to see how many attributes and samples there are in each.

```python
# Your code for Exercise 1.1
"""Select Y as LC50, the value to predict, and X are the other columns"""
```

```python
# Your code for Exercise 1.2
from sklearn.model_selection import train_test_split

# Diabetes dataset:
diabetes_train_set, diabetes_test_set = train_test_split(diabetes_df, test_size=0.2, random_state=42)
diabetes_train_set, diabetes_val_set = train_test_split(diabetes_train_set, test_size=0.25, random_state=42)

y_diabetes = diabetes_train_set["Outcome"]
x_diabetes = diabetes_train_set.drop("Outcome", axis=1)

x_diabetes_test = diabetes_test_set.drop("Outcome", axis=1)
y_diabetes_test = diabetes_test_set["Outcome"]

print(f"Diabetes_df: There are {len(diabetes_train_set)} instances in the training set, {len(diabetes_val_set)} in the validation set, and {len(diabetes_test_set)} in the test set")

# aquatic_toxicity dataset:
aquatic_toxicity_train_set, aquatic_toxicity_test_set = train_test_split(aquatic_toxicity_df, test_size=0.2, random_state=42)
aquatic_toxicity_train_set, aquatic_toxicity_val_set = train_test_split(aquatic_toxicity_train_set, test_size=0.25, random_state=42)

y_aquatic_toxicity = aquatic_toxicity_train_set["LC50"]
x_aquatic_toxicity = aquatic_toxicity_train_set.drop("LC50", axis=1)

x_aquatic_toxicity_test = aquatic_toxicity_test_set.drop("LC50", axis=1)
y_aquatic_toxicity_test = aquatic_toxicity_test_set["LC50"]

print(f"aquatic_toxicity_df: There are {len(aquatic_toxicity_train_set)} instances in the training set, {len(aquatic_toxicity_val_set)} in the validation set, and {len(aquatic_toxicity_test_set)} in the test set")
```

```python
# Your code for Exercise 1.3
print("y_diabetes:", y_diabetes.shape)
print("x_diabetes:", x_diabetes.shape)
print("diabetes_val_set:",diabetes_val_set.shape)
print("diabetes_test_set:",diabetes_test_set.shape)
print("")
print("y_aquatic_toxicity",y_aquatic_toxicity.shape)
print("x_aquatic_toxicity",x_aquatic_toxicity.shape)
print("aquatic_toxicity_val_set",aquatic_toxicity_val_set.shape)
print("aquatic_toxicity_test_set",aquatic_toxicity_test_set.shape)
```

### Exercise 2: Training your first models
1. Train `LinearRegression` on the regression dataset (Aquatic Toxicity).
2. Train `LogisticRegression` on the classification dataset (Diabetes).
3. Predict on the test set for each.

```python
# Your code for Exercise 2.1
# Training of a linear regression
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_aquatic_toxicity, y_aquatic_toxicity)
```

```python
# Your code for Exercise 2.2
# Your code for Exercise 2.1
# Training of a linear regression
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=10000, tol=0.0001)
log_reg.fit(x_diabetes, y_diabetes)
```

```python
# Your code for Exercise 2.3
import sklearn

# Comparem les Y generades amb les Y reals, per mirar com de bé funciona el model de diabetes
array_predict_diabetes = log_reg.predict(x_diabetes_test)

# Comparem les Y generades amb les Y reals, per mirar com de bé funciona el model de toxicity
array_predict_aquatic_toxicity = lin_reg.predict(x_aquatic_toxicity_test)
```

```python
# Your code for Exercise 2.4
print(array_predict_diabetes[0:5])
print(array_predict_aquatic_toxicity[0:5])
```

### Exercise 3: Evaluate your models
1. For regression, use the following metrics to check how good your predictions were: `mean_squared_error`, `r2_score`.
2. Do the same for classification for these: `accuracy_score`, `confusion_matrix`
3. Interpret: Are predictions close? For classification, is there a clear tendency in the model (too many false positives, false negatives, etc.)?

```python
# Your code for Exercise 3.1

from sklearn.metrics import mean_squared_error, r2_score
mean_square = mean_squared_error(y_aquatic_toxicity_test, array_predict_aquatic_toxicity)
r2_score_value = r2_score(y_aquatic_toxicity_test, array_predict_aquatic_toxicity)
print("mean_square:", mean_square)
print("r2_score:", r2_score_value)

# Print the relation between real values and predicted values
import matplotlib.pyplot as plt
plt.figure(0)
plt.scatter(array_predict_aquatic_toxicity, y_aquatic_toxicity_test)
X = np.linspace(0,7,10)
plt.plot(X,X)
plt.show()
```

```python
# Your code for Exercise 3.2

# Accuracy score
accuracy = (array_predict_diabetes == y_diabetes_test).mean()
print(f"Accuracy: {accuracy}")

# See confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

conf_mat_diabetes = confusion_matrix(y_diabetes_test, array_predict_diabetes)
display_labels = ["","","",""]
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat_diabetes)
disp.plot()
```

### Exercise 4: Compare multiple models
1. For the regression problem: Train and compare `LinearRegression`(already done), `DecisionTreeRegressor`, and `KNeighborsRegressor`.
2. For the classification problem: Train and compare `LogisticRegression` (already done), `KNeighborsClassifier`, and `DecisionTreeClassifier`.
3. Evaluate using R² score and accuracy each, respectively.
4. Which model performs best in each case? Why do you think that might be? Try to understand how the models work. Here is the documentation for each of them:
  - [LinearRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)
  - [DecisionTreeRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html)
  - [KNeighborRegressor](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html)
  - [LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)
  - [KNeighborsClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)
  - [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html)

```python
# Your code for Exercise 4.1
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

tree_classifier = DecisionTreeClassifier()
tree_classifier.fit(x_diabetes, y_diabetes)
tree_prediction_classifier = tree_classifier.predict(x_diabetes_test)

neighbor_classifier = KNeighborsClassifier()
neighbor_classifier.fit(x_diabetes, y_diabetes)
neighbor_prediction_classifier = neighbor_classifier.predict(x_diabetes_test)
```

```python
# Your code for Exercise 4.2
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

tree_regressor = DecisionTreeRegressor()
tree_regressor.fit(x_aquatic_toxicity, y_aquatic_toxicity)
tree_prediction = tree_regressor.predict(x_aquatic_toxicity_test)

neighbor_regressor = KNeighborsRegressor(n_neighbors=10)
neighbor_regressor.fit(x_aquatic_toxicity, y_aquatic_toxicity)
neighbor_prediction = neighbor_regressor.predict(x_aquatic_toxicity_test)
```

```python
# Your code for Exercise 4.3
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

mean_square_tree = mean_squared_error(y_aquatic_toxicity_test, tree_prediction)
r2_score_value_tree = r2_score(y_aquatic_toxicity_test, tree_prediction)
print("mean_square_tree:", mean_square_tree)
print("r2_score_tree:", r2_score_value_tree)

# Print the relation between real values and predicted values
import matplotlib.pyplot as plt
plt.figure(0)
plt.scatter(tree_prediction, y_aquatic_toxicity_test)
X = np.linspace(0,7,10)
plt.plot(X,X)
plt.show()

mean_square_neighbor = mean_squared_error(y_aquatic_toxicity_test, neighbor_prediction)
r2_score_value_neighbor = r2_score(y_aquatic_toxicity_test, neighbor_prediction)
print("mean_square_neighbor:", mean_square_neighbor)
print("r2_score_neighbor:", r2_score_value_neighbor)

# Print the relation between real values and predicted values
import matplotlib.pyplot as plt
plt.figure(1)
plt.scatter(neighbor_prediction, y_aquatic_toxicity_test)
X = np.linspace(0,7,10)
plt.plot(X,X)
plt.show()

# Accuracy score
accuracy = (tree_prediction_classifier == y_diabetes_test).mean()
print(f"Accuracy tree classifier: {accuracy}")

# See confusion matrix
conf_mat_diabetes_tree = confusion_matrix(y_diabetes_test, tree_prediction_classifier)
display_labels = ["","","",""]
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat_diabetes_tree)
disp.plot()

# Accuracy score
accuracy = (neighbor_prediction_classifier == y_diabetes_test).mean()
print(f"Accuracy neighbor classifier: {accuracy}")

# See confusion matrix
conf_mat_diabetes_neighbor = confusion_matrix(y_diabetes_test, neighbor_prediction_classifier)
display_labels = ["","","",""]
disp = ConfusionMatrixDisplay(confusion_matrix=conf_mat_diabetes_neighbor)
disp.plot()
```

### Exercise 7: Hyperparameter tuning using Grid Search

1. Use `GridSearchCV` with `LogisticRegression`. Try this grid: `C=[0.01, 0.1, 1, 10]`, `penalty=['l1', 'l2']`, `solver='liblinear'`.
2. You can also run a cross validation with this function. Try a `cv=5`.
3. Print what the best parameters and score were.

`GridSearchCV` documentation: [here](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html).

```python
# Your code for Exercise 7.1
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

# Cargamos un dataset de ejemplo (puedes usar el tuyo)
# X, y = load_breast_cancer(return_X_y=True)
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Definimos el modelo
model = LogisticRegression(solver='liblinear', max_iter=10000)

# Definimos el grid de hiperparámetros
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2']
}

# Creamos el GridSearch con validación cruzada de 5 particiones
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
```

```python
# Your code for Exercise 7.2
grid_search.fit(x_diabetes, y_diabetes)
```

```python
# Your code for Exercise 7.3
print("Best parameters:", grid_search.best_params_)
print("Best cross-validated accuracy score:", grid_search.best_score_)
```

### Exercise 8: Hyperparameter tuning using Random Search

1. Use `RandomizedSearchCV` on the [`Ridge`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.Ridge.html) model (for regression). Sample only one parameter: `alpha` from `scipy.stats.uniform(0.01, 10)`.
2. Limit the random search to only 10 iterations.
3. Print the best alpha and score.

```python
# Your code for Exercise 8.1
from sklearn.linear_model import Ridge
from sklearn.model_selection import RandomizedSearchCV
from sklearn.datasets import load_diabetes
from scipy.stats import uniform
from sklearn.model_selection import train_test_split

# Cargar datos de ejemplo (puedes usar los tuyos)
# X, y = load_diabetes(return_X_y=True)
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Modelo base
model = Ridge()

# Distribución de alpha
param_dist = {
    'alpha': uniform(0.01, 10)  # alpha ∈ [0.01, 10.01)
}

# RandomizedSearchCV con 10 combinaciones
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=10,
    scoring='r2',   # Porque es regresión
    cv=5,
    random_state=42,
    n_jobs=-1
)
```

```python
# Your code for Exercise 8.2
random_search.fit(x_diabetes, y_diabetes)
```

```python
# Your code for Exercise 8.3
print("Best alpha:", random_search.best_params_['alpha'])
print("Best cross-validated R² score:", random_search.best_score_)
```
